```python
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(10,110,10)
y=np.arange(12,132,12)
y1=np.arange(15,165,15)
y2=np.arange(20,220,20)
print('X Coridinate',x)
print('Y ordinate',y)
plt.plot(x,y,marker="s",color='k',label='2012',linewidth=2,mec='#CD5C5C',mew=2)
plt.plot(x,y1,marker="o",color='g',label='2016',lw=2)
plt.plot(x,y2,marker=">",color="b",label='2017',lw=2,ls='dotted')
plt.xlabel("X cordinate")
plt.ylabel("Y Cordinates")
plt.title("Test data 1")
plt.legend(ncol=3)
plt.show()
```

# BAR PLOT

```python
fig=plt.figure()
axis=fig.add_axes([0,0,1,1])
cities=['ASSAM','GOA','DELHI','GUJRAT','BIHAR']
c=['lightblue','lightgreen','red','green','m']
cases=[20000,3000,102000,40000,23600]
axis.bar(cities,cases,color=c)
axis.set_xlabel('Cities_Nmae')
axis.set_ylabel('no _of_cases')
axis.set_title("CITY CORORNA CASES")
plt.show()
```

```python
fig=plt.figure()
axis=fig.add_axes([0,0,1,1])
cities=['ASSAM','GOA','DELHI','GUJRAT','BIHAR']
c=['lightblue','lightgreen','red','green','m']
cases=[20000,3000,102000,40000,23600]
axis.barh(cities,cases,color=c)
axis.set_ylabel('Cities_Nmae')
axis.set_xlabel('no _of_cases')
axis.set_title("CITY CORORNA CASES")
plt.show()
```

```python
data=[[30,25,50,30],[40,23,5,17],[35,22,45,19]]
x=np.arange(4)
fig=plt.figure(figsize=(5,5))
axis=fig.add_axes([0,0,1,1])
axis.bar(x,data[0],color='r',width=0.30)
axis.bar(x+0.30,data[1],color='green',width=0.30)
axis.bar(x+0.60,data[2],color='blue',width=0.30)
```

```python
oil=[20,35,30,35,40]
gas=[25,32,34,20,25]
ind=np.arange(5)
fig=plt.figure()
axis=fig.add_axes([0,0,1,1])
axis.bar(ind,oil,width=0.40,color='y')
axis.bar(ind,gas,width=0.40,color='lightblue',bottom=oil)
axis.set_xlabel('data')
axis.set_ylabel('indicators')
axis.legend(labels=['OIL','GAS'])
plt.show()
```
